```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/SaYoPillow.csv")
data.head()
```

```python
data.columns=['taxa_ronco', 'taxa_respiração', 'temperatura_corporal', 'movimento_membro', 'oxigenio_sangue',
             'movimento_dos_olhos', 'horas_de_sono', 'frequencia_cardiaca', 'nivel_de_stress']
```

```python
data.columns
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
sns.pairplot(data)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='taxa_ronco',y='nivel_de_stress',data=data)
plt.xlabel("taxa_ronco")
plt.ylabel('nivel-de_stress')
plt.title('Taxa de Ronco vs Nível de Stress')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='taxa_respiração',y='nivel_de_stress',data=data)
plt.xlabel("taxa_respiração")
plt.ylabel('nivel_de_stress')
plt.title('Taxa de Respiração vs Nível de Stress')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='temperatura_corporal',y='nivel_de_stress',data=data)
plt.xlabel("temperatura_corporal")
plt.ylabel('nivel_de_stress')
plt.title('Temperatura Corporal vs Nível de Stress')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='oxigenio_sangue',y='nivel_de_stress',data=data)
plt.xlabel("oxigenio_sangue")
plt.ylabel('nivel_de_stress')
plt.title('Oxigênio no Sangue vs Nível de Stress')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='movimento_dos_olhos',y='nivel_de_stress',data=data)
plt.xlabel("movimento_dos_olhos")
plt.ylabel('nivel_de_stress')
plt.title('Movimento dos olhos vs Nível de Stress')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='horas_de_sono',y='nivel_de_stress',data=data)
plt.xlabel("horas_de_sono")
plt.ylabel('nivel_de_stress')
plt.title('Horas de Sono vc Nível de Stress')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='frequencia_cardiaca',y='nivel_de_stress',data=data)
plt.xlabel("frequencia_cardiaca")
plt.ylabel('nivel_de_stress')
plt.title('Frequência Cardiaca vs Nível de Stress')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10,10))
tc = data.corr()
sns.heatmap(tc,annot=True)
plt.title('Heatmap')
plt.show()
```
